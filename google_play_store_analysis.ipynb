```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('google-play-store-11-2018.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
#Data Cleaning
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df.sort_values('min_installs', ascending=False)
```

```python
df['total_no_reviews'] = df.apply(lambda x: x.rating_one_star + x.rating_two_star + x.rating_three_star + x.rating_four_star + x.rating_five_star,axis = 1)
df.sort_values('total_no_reviews',ascending=False)
```

```python
df.drop(df[df['total_no_reviews'] < 500].index,inplace = True)
```

```python
df.drop('total_no_reviews',axis = 1,inplace = True)
```

```python
df
```

```python
df_hidden = df.loc[:, df.columns != 'app_id']
df_hidden
```

```python
df_filtered = (df_hidden.loc[(df_hidden.ad_supported == True) & (df_hidden.min_installs > 100000)]
               .sort_values(by="min_installs", ascending=False))
```

```python
df_filtered_grouped = df_filtered.groupby([df_filtered.genre]).size().sort_values(ascending=False)
```

```python
# Set the figure size
plt.figure(figsize=(9,7))

# Create a barplot using Seaborn
top_10_values = df_filtered_grouped.head(10)
sns.barplot(x=top_10_values.index, y=top_10_values.values, hue = top_10_values.index, palette="magma",legend = False)

# Add labels and title
plt.xlabel('Number of Apps')
plt.ylabel('Genre')
plt.title('Number of Apps by Genre with over 100K Downloads and Ad Support')
plt.xticks(rotation = 20)
# Show the plot
plt.tight_layout()
plt.show()

# Save the plot as a PNG image
plt.savefig('apps_by_genre.png')
```
